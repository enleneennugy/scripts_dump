```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
    
df=pd.read_csv('missing_data.csv', sep='\t',  index_col = 0)
df = df.fillna(0)    
cmap = "RdBu_r"
sns.set(font_scale=0.8)
g=sns.clustermap(df,
                cmap="Blues",  
                row_cluster=True,
                col_cluster=True,
                figsize=(50, 130))
plt.setp(g.ax_heatmap.yaxis.get_majorticklabels(),fontstyle='italic',rotation=0)
plt.xlabel('xlabel', fontsize=18)

plt.xlabel('percentage', rotation=0)
plt.yticks(rotation=0)

#compatible illustration fond (true type)
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
g.savefig("missing_data_blue.pdf")
```
